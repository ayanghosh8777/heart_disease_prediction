```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df=pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

## LINEAR SVC/LINEAR KERNEL( accuracy: 81 %)

```python
svc=SVC(kernel='linear')
svc.fit(x_train_scaled,y_train)
```

```python
y_pred=svc.predict(x_test_scaled)
acc=accuracy_score(y_test,y_pred)
```

```python
## PERFORMANCE METRICES
print(f"THE ACCURACY:{acc}")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
svc.support_vectors_
```

```python
svc.dual_coef_
```

```python
svc.n_support_
```

## POLYNOMIAL KERNEL( accuracy: 91 %)

```python
poly=SVC(kernel='poly',probability=True)
poly.fit(x_train_scaled,y_train)
```

```python
y_pred=poly.predict(x_test_scaled)
acc=accuracy_score(y_test,y_pred)
```

```python
## PERFORMANCE METRICES
print(f"THE ACCURACY:{acc}")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
poly.support_vectors_
```

```python
poly.n_support_
```

## RBF KERNEL( accuracy: 89 %)

```python
rbf=SVC(kernel='rbf')
rbf.fit(x_train_scaled,y_train)
```

```python
y_pred=rbf.predict(x_test_scaled)
acc=accuracy_score(y_test,y_pred)
```

```python
## PERFORMANCE METRICES
print(f"THE ACCURACY:{acc}")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
rbf.support_vectors_
```

```python
rbf.n_support_
```

```python
from sklearn.model_selection import cross_val_score
```

```python
linear_cv = cross_val_score(svc, x_train_scaled, y_train, cv=5)
rbf_cv = cross_val_score(rbf, x_train_scaled, y_train, cv=5)
poly_cv = cross_val_score(poly, x_train_scaled, y_train, cv=5)

print(linear_cv.mean())
print(rbf_cv.mean())
print(poly_cv.mean())
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'C':[0.1,1,10,100],
    'degree':[2,3,4,5],
    'gamma':['scale','auto'],
    'coef0':[0,1,10,100]
}
grid=GridSearchCV(estimator=poly,param_grid=param_grid,scoring='accuracy',n_jobs=-1,cv=cv)
```

```python
grid.fit(x_train_scaled,y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
best_model=grid.best_estimator_
y_pred = best_model.predict(x_test_scaled)
```

```python
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```
